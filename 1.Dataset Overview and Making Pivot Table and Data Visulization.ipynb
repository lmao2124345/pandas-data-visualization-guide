```python
import pandas as pd 
```

```python
df_population_raw= pd.read_csv('population_total.csv',encoding='latin-1')
```

```python
df_population_raw
```

# Making a Pivot Table 

```python
#dropping null values 
df_population_raw.dropna(inplace=True)
```

```python
# making a pivt table 
df_pivot=df_population_raw.pivot(index='year',
                        columns='country',
                        values='population')
```

```python
# selecting some countries
df_pivot=df_pivot[['United States','India','China','Indonesia','Brazil']]
```

```python
df_pivot
```

# Lineplot

```python
#importing matplotlib
import matplotlib.pyplot as plt
df_pivot.plot(kind='line',
              xlabel='Year',
              ylabel='Population',
              title='Population(1955-2020',
              figsize=(8,4))
#save plot 
plt.savefig('my_test.png')
#plt.show()
plt.show()
```

```python
df_pivot.to_excel('pivot_table.xlsx')
```

# Barplot

## Single Barplot

```python
# selecting only one year (2020)
df_pivot_2020=df_pivot[df_pivot.index.isin([2020])]
```

```python
df_pivot_2020
```

```python
# transpose datframe from rows to columns or viceversa 
df_pivot_2020=df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind='bar',
                  color='pink',
                  xlabel='Year',
                  ylabel='Population',
                  title='Population 2020')
```

## Barplot grouped by "n" variables 

```python
# selecting a few years 
df_pivot_sample=df_pivot[df_pivot.index.isin([1980,1990,200,2010,2020])]
```

```python
#making grouped barplot
df_pivot_sample.plot(kind='bar',
                     xlabel='Year',
                  ylabel='Population',)
```

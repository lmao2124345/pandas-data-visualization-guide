```python
import pandas as pd 
import cufflinks as cf
import plotly.express as px

from IPython.display import display, HTML

cf.set_config_file(sharing='public', theme='ggplot',offline=True)
```

```python
cf.getThemes()
```

```python
df_population_raw=pd.read_csv('population_total.csv')
```

```python
#dropping null values 
df_population_raw.dropna(inplace=True)

# making a pivt table 
df_pivot=df_population_raw.pivot(index='year',
                        columns='country',
                        values='population')

# selecting some countries
df_pivot=df_pivot[['United States','India','China','Indonesia','Brazil']]

df_pivot
```

# Lineplot

```python
pd.options.plotting.backend = "plotly" 
df_pivot.plot(kind='line')
```

# Barplot

```python
# selecting only one year (2020)
df_pivot_2020=df_pivot[df_pivot.index.isin([2020])]
# transpose datframe from rows to columns or viceversa 
df_pivot_2020=df_pivot_2020.T
```

```python
df_pivot_2020.plot(kind='bar')
```

## Barplot grounded by "n" variables 

```python
# selecting few years 
df_pivot_sample=df_pivot[df_pivot.index.isin([1980,1990,200,2010,2020])]
```

```python
df_pivot_sample.plot(kind='bar')
```

# Piechart

```python
# chaning column name 
df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
```

```python
#making the plot 
px.pie(df_pivot_2020, values='2020', names=df_pivot_2020.index, title='Population Distribution 2020')
```
